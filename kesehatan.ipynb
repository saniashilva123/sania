```python
import pandas as pd 
path= '/content/drive/MyDrive/csv/kesehatan - Sheet1.csv'
df = pd.read_csv(path)


df.describe()
```

```python
df
```

```python
import matplotlib.pyplot as plt

df.plot(kind='scatter', x='Jumlah Balita Terdaftar',y= 'Jumlah Balita Aktif')

plt.title('Indeks kesejahteraan warga')
plt.xlabel('Jumlah Balita Terdaftar')
plt.ylabel('Jumlah Balita Aktif')
plt.xlim(30, 90)
plt.ylim(30, 90)
plt.grid(True)
plt.show()
```

```python

import numpy as np
X = np.array(df['Jumlah Balita Terdaftar'])
y = np.array(df['Jumlah Balita Aktif'])

print(f'X : {X}')
print(f'y : {y}')
```

```python
X = X.reshape(-1, 1)
X.shape
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
```

```python

X_vis =np.array([0, 90]).reshape(-1, 1)
y_vis = model.predict(X_vis)
```

```python
plt.scatter(X, y)
plt.plot(X_vis, y_vis, 'r')
plt.title('kesehatan')
plt.xlabel('Jumlah Balita Terdaftar')
plt.ylabel('Jumlah Balita Aktif')
plt.xlim(30, 90)
plt.ylim(30, 90 )
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
pendidikan = df['Jumlah Balita Terdaftar']
hasil = df['Jumlah Balita Aktif' ]
c = ['red', 'green', 'blue']
plt.title("pendidikan")
plt.bar (pendidikan, hasil,  color = c)
plt.show()
```
