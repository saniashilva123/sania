```python
import pandas as pd 
path= '/content/drive/MyDrive/csv/lingkungan hidup - indeks-standar-pencemar-udara-di-provinsi-dki-jakarta-bulan-maret-tahun-2020.csv'
df = pd.read_csv(path)


df.describe()
```

```python
df
```

```python
import matplotlib.pyplot as plt

df.plot(kind='scatter', x='o3',y = 'max')

plt.title('Indeks kesejahteraan warga')
plt.xlabel('o3')
plt.ylabel('max')
plt.xlim(30, 90)
plt.ylim(30, 90)
plt.grid(True)
plt.show()
```

```python

import numpy as np
X = np.array(df['o3'])
y = np.array(df['max'])

print(f'X : {X}')
print(f'y : {y}')
```

```python
X = X.reshape(-1, 1)
X.shape
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
```

```python

X_vis =np.array([0, 90]).reshape(-1, 1)
y_vis = model.predict(X_vis)
```

```python
plt.scatter(X, y)
plt.plot(X_vis, y_vis, 'r')
plt.title('kesehatan')
plt.xlabel('o3')
plt.ylabel('max')
plt.xlim(30, 90)
plt.ylim(30, 90)
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
pendidikan = df['o3']
hasil = df['max' ]
c = ['red', 'green', 'blue']
plt.title("lingkungan")
plt.bar (pendidikan, hasil,  color = c)
plt.show()
```

```python
import matplotlib.pyplot as plt
#pie chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
plt.title("pendidikan")
usia = df['o3']
hasil = df['max' ]
ax.pie(hasil,labels=usia,autopct='%1.2f%%')
plt.show()
```
