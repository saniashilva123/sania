```python
import pandas as pd 
path= '/content/drive/MyDrive/csv/tahun-20172018-ganjil---data-jumlah-sekolah-dasar-di-kecamatan-andir.csv'
df = pd.read_csv(path)


df.describe()
```

```python
df
```

```python
import matplotlib.pyplot as plt

df.plot(kind='scatter', x='Peserta Didik',y= 'Rombongan Belajar')

plt.title('Indeks kesejahteraan warga')
plt.xlabel('Peserta Didik')
plt.ylabel('Rombongan Belajar')
plt.xlim(0, 1000)
plt.ylim(0, 33)
plt.grid(True)
plt.show()
```

```python

import numpy as np
X = np.array(df['Peserta Didik'])
y = np.array(df['Rombongan Belajar'])

print(f'X : {X}')
print(f'y : {y}')
```

```python
X = X.reshape(-1, 1)
X.shape
```

```python
X
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X, y)
```

```python

X_vis =np.array([0, 1000]).reshape(-1, 1)
y_vis = model.predict(X_vis)
```

```python
plt.scatter(X, y)
plt.plot(X_vis, y_vis, 'r')
plt.title('indeks kesejahteraan warga')
plt.xlabel('Peserta Didik')
plt.ylabel('Rombongan Belajar')
plt.xlim(0, 1000)
plt.ylim(0, 33 )
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
pendidikan = df['Peserta Didik']
hasil = df['Rombongan Belajar' ]
c = ['red', 'green', 'blue']
plt.title("pendidikan")
plt.bar (pendidikan, hasil,  color = c)
plt.show()
```

```python
import matplotlib.pyplot as plt
#pie chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
plt.title("pendidikan")
usia = df ['Peserta Didik']
hasil = df ['Rombongan Belajar']
ax.pie(hasil,labels=usia,autopct='%1.2f%%')
plt.show()
```
